The most common use for querying is to read data from our tables. We can do this using a `SELECT ... FROM` statement<br>

However, running a `SELECT ... FROM` statement can cause issues when the data set is extensive. If you want to examine the data, use `LIMIT n` to instruct **Trino to retrieve only the first n rows**.<br>

We can use the `‘WHERE’ clause` to retrieve rows that match specific criteria. We can specify one or more filters within the`’ WHERE’ clause`. The WHERE clause with more than one filter can use combinations of `AND and OR` criteria to combine the filter criteria, as shown below.

````python
%%sql
-- all customer rows that have c_nationkey = 20 and c_acctbal > 1000
SELECT
  *
FROM
  customer
WHERE
  c_nationkey = 20
  AND c_acctbal > 1000
LIMIT
  10;
````

````python
%%sql
-- All customer rows that have (c_nationkey = 20 and c_acctbal > 1000) or rows that have c_nationkey = 11
SELECT
  *
FROM
  customer
WHERE
  (
    c_nationkey = 20
    AND c_acctbal > 1000
  )
  OR c_nationkey = 11
LIMIT
  10;
````

##  SQL Joins Recap

````python
%%sql
SELECT
  COUNT(o.o_orderkey),
  COUNT(l.l_orderkey)
FROM
  orders o
  JOIN lineitem l ON o.o_orderkey = l.l_orderkey
  AND o.o_orderdate BETWEEN l.l_shipdate - INTERVAL '5' DAY AND l.l_shipdate  + INTERVAL '5' DAY;
````

````python
%%sql
SELECT
  COUNT(o.o_orderkey) AS order_rows_count,
  COUNT(l.l_orderkey) AS lineitem_rows_count
FROM
  orders o
  LEFT JOIN lineitem l ON o.o_orderkey = l.l_orderkey
  AND o.o_orderdate BETWEEN l.l_shipdate - INTERVAL '5' DAY AND l.l_shipdate  + INTERVAL '5' DAY;
````

The output will include all rows from orders and the rows from lineitem that were able to find at least one matching row from the orders table with the specified join condition (same orderkey and orderdate within a 5-day window of the ship date).

We can also see that the number of rows from the orders table is 151933 & from the lineitem table is 24613. The number of rows in orders is 150000, but the join condition produces 151933 since some orders match with multiple line items.

````python
%%sql
SELECT
  COUNT(o.o_orderkey) AS order_rows_count,
  COUNT(l.l_orderkey) AS lineitem_rows_count
FROM
  orders o
  FULL OUTER JOIN lineitem l ON o.o_orderkey = l.l_orderkey
  AND o.o_orderdate BETWEEN l.l_shipdate - INTERVAL '5' DAY AND l.l_shipdate  + INTERVAL '5' DAY;
````

### Using Subqueries

Create a report that shows the nation, how many items it supplied (by suppliers in that nation), and how many items it purchased (by customers in that nation).

The query below answers the question: how many items it supplied (by suppliers in that nation)

````python
%%sql
SELECT
  n.n_nationkey,
  SUM(l.l_quantity) AS quantity
FROM
  lineitem l
  JOIN supplier s ON l.l_suppkey = s.s_suppkey
  JOIN nation n ON s.s_nationkey = n.n_nationkey
GROUP BY
  n.n_nationkey;
````

The below query answers the question: how many items it purchased (by customers in that nation)

````python
%%sql
SELECT
  n.n_nationkey,
  SUM(l.l_quantity) AS quantity
FROM
  lineitem l
  JOIN orders o ON l.l_orderkey = o.o_orderkey
  JOIN customer c ON o.o_custkey = c.c_custkey
  JOIN nation n ON c.c_nationkey = n.n_nationkey
GROUP BY
  n.n_nationkey;
````

````python
%%sql
SELECT
  n.n_name AS nation_c_name,
  s.quantity AS supplied_items_quantity,
  c.quantity AS purchased_items_quantity
FROM
  nation n
  LEFT JOIN (
    SELECT
      n.n_nationkey,
      SUM(l.l_quantity) AS quantity
    FROM
      lineitem l
      JOIN supplier s ON l.l_suppkey = s.s_suppkey
      JOIN nation n ON s.s_nationkey = n.n_nationkey
    GROUP BY
      n.n_nationkey
  ) s ON n.n_nationkey = s.n_nationkey
  LEFT JOIN (
    SELECT
      n.n_nationkey,
      SUM(l.l_quantity) AS quantity
    FROM
      lineitem l
      JOIN orders o ON l.l_orderkey = o.o_orderkey
      JOIN customer c ON o.o_custkey = c.c_custkey
      JOIN nation n ON c.c_nationkey = n.n_nationkey
    GROUP BY
      n.n_nationkey
  ) c ON n.n_nationkey = c.n_nationkey;
````

To stack tables on top of each other - use UNION or UNION ALL operator. UNION removes duplicate rows and UNION ALL does not remove duplicate rows.

Use `EXCEPT` to subtract tables - When you want to retrieve all the rows from the first table that are not present in the second table

````python
%%sql
SELECT
  c_custkey,
  c_name
FROM
  customer
WHERE
  c_name LIKE '%_91%'
EXCEPT
SELECT
  c_custkey,
  c_name
FROM
  customer
WHERE
  c_name LIKE '%_91%'
````

The above query has no results as we are subtracting a table with itself so all values from the first table will be present in the second table too hence the query will result in zero records.

````python
%%sql
CREATE VIEW
  nation_supplied_purchased_quantity AS
SELECT
  n.n_name AS nation_c_name,
  s.quantity AS supplied_items_quantity,
  c.quantity AS purchased_items_quantity
FROM
  nation n
  LEFT JOIN (
    SELECT
      n.n_nationkey,
      SUM(l.l_quantity) AS quantity
    FROM
      lineitem l
      JOIN supplier s ON l.l_suppkey = s.s_suppkey
      JOIN nation n ON s.s_nationkey = n.n_nationkey
    GROUP BY
      n.n_nationkey
  ) s ON n.n_nationkey = s.n_nationkey
  LEFT JOIN (
    SELECT
      n.n_nationkey,
      SUM(l.l_quantity) AS quantity
    FROM
      lineitem l
      JOIN orders o ON l.l_orderkey = o.o_orderkey
      JOIN customer c ON o.o_custkey = c.c_custkey
      JOIN nation n ON c.c_nationkey = n.n_nationkey
    GROUP BY
      n.n_nationkey
  ) c ON n.n_nationkey = c.n_nationkey;
````

Write a query that shows the number of items returned for each region name

````python
%%sql
SELECT
  r.r_name AS region_name,
  COUNT(l.l_returnflag) AS items_returned
FROM
  region r
  JOIN nation n ON r.r_regionkey = n.n_nationkey
  JOIN customer c ON n.n_nationkey = c.c_nationkey
  JOIN orders o ON c.c_custkey = o.o_custkey
  JOIN lineitem l ON o.o_orderkey = l.l_orderkey
WHERE
  l.l_returnflag = 'R'
GROUP BY
  r.r_name;
````

List the top 10 most selling parts (part name)

````python
%%sql
SELECT
  p.p_name AS part_name,
  CAST(SUM(l.l_quantity) AS INT) AS total_units_sold
FROM
  part p
  JOIN lineitem l ON p.p_partkey = l.l_partkey
GROUP BY
  1
ORDER BY
  2 DESC
LIMIT
  10
````

Sellers (name) who have sold at least one of the top 10 selling parts

````python
%%sql
SELECT DISTINCT
  s.s_name
FROM
  supplier s
  JOIN lineitem l ON s.s_suppkey = l.l_suppkey
WHERE
  l.l_partkey IN (
    SELECT
      l.l_partkey
    FROM
      lineitem l
    GROUP BY
      1
    ORDER BY
      (SUM(l.l_quantity)) DESC
    LIMIT
      10
  )
````

        ELSE 'low'

Average time (in days) between receiptdate and shipdate for each nation (name)

````python
%%sql
SELECT
  n.n_name,
  AVG(l.l_receiptdate - l.l_shipdate) AS avg_shipping_delay_days
FROM
  lineitem l
  JOIN supplier s ON l.l_suppkey = s.s_suppkey
  JOIN nation n ON s.s_nationkey = n.n_nationkey
GROUP BY
  n.n_name
````

````python
%%sql
WITH
  lineitem_delay AS (
    SELECT
      l.l_suppkey,
      l.l_receiptdate - l.l_shipdate AS shipping_delay_days
    FROM
      lineitem l
  ),
  lineitem_with_nation AS (
    SELECT
      ld.shipping_delay_days,
      n.n_name AS nation_name
    FROM
      lineitem_delay ld
      JOIN supplier s ON ld.l_suppkey = s.s_suppkey
      JOIN nation n ON s.s_nationkey = n.n_nationkey
  )
SELECT
  nation_name,
  AVG(shipping_delay_days) AS avg_shipping_delay_days
FROM
  lineitem_with_nation
GROUP BY
  nation_name;
````

Calculate the money lost due to discounts. Use the `lineitem` to retrieve the prices of items (excluding discounts) that are part of an order and compare them to the order.

The `l_extendedprice` column does not include discounts & the `o_totalprice` column includes discounts.

````python
%%sql
WITH
  order_item_totals AS (
    SELECT
      l.l_orderkey,
      SUM(l_extendedprice) AS total_item_price
    FROM
      lineitem l
    GROUP BY
      l.l_orderkey
  )
SELECT
  o.o_orderkey,
  o.o_totalprice,
  oit.total_item_price - o.o_totalprice AS discount_loss
FROM
  order_item_totals oit
  JOIN orders o ON oit.l_orderkey = o.o_orderkey
````

Sellers (name) who have sold at least one of the top 10 selling parts (use CTE)

````python
%%sql
WITH
  top_10_selling_parts AS (
    SELECT
      l.l_partkey
    FROM
      lineitem l
    GROUP BY
      l.l_partkey
    ORDER BY
      (SUM(l.l_quantity)) DESC
    LIMIT
      10
  )
SELECT DISTINCT
  s.s_name
FROM
  supplier s
  JOIN lineitem l ON s.s_suppkey = l.l_suppkey
  JOIN top_10_selling_parts t ON l.l_partkey = t.l_partkey
````

## Window Functions
Window functions allow you to operate on a set of rows at a time and produce output that has the exact grain as the input (vs GROUP BY, which operates on a set of rows, but also changes the meaning of an output row).

Common scenarios when you want to use window functions:

- Calculate running metrics/sliding window over rows in the table (aggregate functions)
- Ranking rows based on values in column(s) (ranking functions)
- Access other rows’ values while operating on the current row (value functions)
- Any combination of the above

````python
%%sql
SELECT
  o_custkey,
  o_orderdate,
  o_totalprice,
  AVG(o_totalprice) -- FUNCTION 
  OVER (
    PARTITION BY
      o_custkey -- PARTITION
    ORDER BY
      o_orderdate -- ORDER BY; ASCENDING ORDER unless specified as DESC
  ) AS running_sum
FROM
  orders
WHERE
  o_custkey = 4
ORDER BY
  o_orderdate
LIMIT
  10;
````

Write a query on the orders table that has the following output:

~~~order_month,
o_custkey,
total_price,
three_mo_total_price_avg
consecutive_three_mo_total_price_avg: The consecutive 3 month average of total_price for that customer. Note that this should only include months that are chronologically next to each other.
~~~
>Hint: Use `CAST(strftime(o_orderdate, '%Y-%m-01') AS DATE)` to cast order_month to date format.

>Hint: Use the `INTERVAL` format shown above to construct the window function to compute `consecutive_three_mo_total_price_avg` column.

````python
%%sql
WITH
  monthly_orders AS (
    SELECT
      o.o_custkey,
      CAST(date_format (o.o_orderdate, 'yyyy-M-01') AS DATE) AS order_month,
      SUM(o.o_totalprice) AS total_price
    FROM
      orders o
    GROUP BY
      1,
      2
  )
SELECT
  o_custkey,
  total_price,
  AVG(total_price) OVER (
    PARTITION BY
      o_custkey
    ORDER BY
      order_month ROWS BETWEEN 2 PRECEDING
      AND CURRENT ROW
  ) AS three_mo_total_price_avg
FROM
  monthly_orders
ORDER BY
  1,
  2
````

### Ranking functions

Get the top 3 spending customers per day

````python
%%sql
WITH
  ranking AS (
    SELECT
      o_orderdate,
      o_totalprice,
      o_custkey,
      RANK() OVER (
        PARTITION BY
          o_orderdate
        ORDER BY
          o_totalprice DESC
      ) AS rnk
    FROM
      orders
  )
SELECT
  *
FROM
  ranking
WHERE
  rnk <= 3
ORDER BY
  o_orderdate
LIMIT
  5;
````

````python
%%sql
-- Let's look at an example showing the difference between RANK, DENSE_RANK and ROW_NUMBER
SELECT
  order_date,
  order_id,
  total_price,
  ROW_NUMBER() OVER (
    PARTITION BY
      order_date
    ORDER BY
      total_price
  ) AS ROW_NUMBER,
  RANK() OVER (
    PARTITION BY
      order_date
    ORDER BY
      total_price
  ) AS RANK,
  DENSE_RANK() OVER (
    PARTITION BY
      order_date
    ORDER BY
      total_price
  ) AS DENSE_RANK
FROM
  (
    SELECT
      '2024-07-08' AS order_date,
      'order_1' AS order_id,
      100 AS total_price
    UNION ALL
    SELECT
      '2024-07-08',
      'order_2',
      200
    UNION ALL
    SELECT
      '2024-07-08',
      'order_3',
      150
    UNION ALL
    SELECT
      '2024-07-08',
      'order_4',
      90
    UNION ALL
    SELECT
      '2024-07-08',
      'order_5',
      100
    UNION ALL
    SELECT
      '2024-07-08',
      'order_6',
      90
    UNION ALL
    SELECT
      '2024-07-08',
      'order_7',
      100
    UNION ALL
    SELECT
      '2024-07-10',
      'order_8',
      100
    UNION ALL
    SELECT
      '2024-07-10',
      'order_9',
      100
    UNION ALL
    SELECT
      '2024-07-10',
      'order_10',
      100
    UNION ALL
    SELECT
      '2024-07-11',
      'order_11',
      100
  ) AS orders
ORDER BY
  order_date,
  ROW_NUMBER;
````

Write a query on the orders table that has the following output:
- o_custkey
- order_month: In YYYY-MM format, use strftime(o_orderdate, ‘%Y-%m’) AS order_month
- total_price: Sum of o_totalprice for that month
- three_mo_total_price_avg: The 3 month (previous, current & next) average of total_price for that customer

````python
%%sql
SELECT
  o_custkey,
  order_month,
  total_price,
  AVG(total_price) OVER (
    PARTITION BY
      o_custkey
    ORDER BY
      order_month ROWS BETWEEN 1 PRECEDING
      AND 1 FOLLOWING
  ) AS three_mo_total_price_avg
FROM
  (
    SELECT
      o_custkey,
      date_format (o_orderdate, 'yyyy-MM') AS order_month,
      SUM(o_totalprice) AS total_price
    FROM
      orders
    GROUP BY
      1,
      2
  )
````

From the orders table get the 3 lowest spending customers per day

````python
%%sql
WITH
  ranking AS (
    SELECT
      o_custkey,
      o_totalprice,
      o_orderdate,
      RANK() OVER (
        PARTITION BY
          o_orderdate
        ORDER BY
          o_totalprice
      ) AS rnk
    FROM
      orders
  )
SELECT
  *
FROM
  ranking
WHERE
  rnk <= 3
LIMIT
  6;
````
